```python
import matplotlib.pyplot as plt
import numpy as np
# this image is taken from https://www.tutorialspoint.com/dip/laplacian_operator.htm
img = plt.imread('images/beach.jpeg')
print(img.shape)
plt.imshow(img)
```

```python
red = img[...,0]
green = img[..., 1]
blue = img[...,2]
gray = 0.21 * red + 0.72 * green + 0.07 * blue
gray_img = gray.astype('uint8')
plt.imshow(gray_img, cmap="gray")
print(gray_img.shape)
```

```python
threshold = 100
threshold_image = gray_img.copy()
less_than = threshold_image <= threshold
greater_than = threshold_image > threshold
threshold_image[less_than] = 0
threshold_image[greater_than] = 1
plt.imshow(threshold_image, cmap="gray")
print("shape")
print(threshold_image.shape)
print("max")
print(np.max(threshold_image))
print("min")
print(np.min(threshold_image))
```
